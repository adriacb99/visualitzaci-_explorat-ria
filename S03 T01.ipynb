```python
import pandas as pd
import numpy as np
import seaborn as sns 

df = pd.read_csv('DelayedFlights.csv')
dff = df["UniqueCarrier"].to_numpy()
```

#### Una variable categòrica (UniqueCarrier)

```python
x = "UniqueCarrier"
plt.figure(figsize=(11,7))
sns.set_style("whitegrid")
plot = sns.countplot(x, data=df, order = df.UniqueCarrier.value_counts().index)
plt.title("Número de vuelos con cada UniqueCarrier")
plt.xlabel("Código UniqueCarrier")
plt.ylabel("Número de vuelos")
plt.show()
```

#### Una variable numèrica (ArrDelay)

```python
dff = df["ArrDelay"].to_numpy()
num_bins = 100
plt.figure(figsize=(11,7))
plt.hist(dff, num_bins)
plt.axvline(df.ArrDelay.mean(), color ="Red")
plt.annotate("Mediana", xy=(2,df.ArrDelay.mean()),xytext=(500,700000), arrowprops=dict(arrowstyle='->'))
plt.show()
```
